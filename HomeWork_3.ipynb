### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.
### Если реновации дома не было, то в новом признаке поставьте 0
* **yr_built** - Год постройки дома
* **yr_renovated** - Год ремонта дома

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.head()
```

```python
df.loc[df['yr_renovated'] == 0, 'delta_renovated'] = 0
df.head()
```

```python
df['delta_renovated'].unique() # проверка на отсутствие отрицательных значений
```

## Задача 3
Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby(['bedrooms'], as_index=False).agg({'price':'mean'}).sort_values('price')
```

```python
df['price'].mean()
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома
* **condition** - Значение от 1 до 5 насколько хорошее состояние

```python
# 1 способ. groupby

df.groupby(['condition']).agg({'price':['min', 'mean', 'max']})
```

```python
# 2 способ. pivot_table

df.pivot_table(index='condition',
               aggfunc={'price':['min', 'mean', 'max']})
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида
* **waterfront** - Есть набержная или нет
* **view** - Значение от 0 до 4 насколько хороший вид

```python
# 1 способ. groupby

df.groupby(['waterfront', 'view']).agg({'id':'count'})
```

```python
# 2 способ. crosstab

pd.crosstab(index= df['waterfront'],
            columns= df['view'])
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?
* **bedrooms** - Кол-во спален
* **floors** - Кол-во этажей

```python
# 1 способ. groupby

df.groupby(['floors', 'bedrooms'], as_index=False).agg({'id':'count'}).rename(columns=({'id':'counts'}))\
    .sort_values('counts', ascending=False).iloc[0]
```

```python
# 2 способ. crosstab

pd.crosstab(index= df['floors'],
            columns= df['bedrooms'])
```

```python
# 3 способ. pivot_table

df.pivot_table(index= ['bedrooms', 'floors'],
               aggfunc= {'id':'count'}).sort_values('id', ascending=False)
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
* **condition** - Значение от 1 до 5 насколько хорошее состояние
* **grade** - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна

```python
# 1 способ.groupby

df.groupby(['condition', 'grade']).agg({'price':'median'})
```

```python
# 2 способ. pivot_table

df.pivot_table(index= 'condition',
               columns= 'grade',
               values= 'price',
               aggfunc= 'median')
```

```python
# 3 способ. crosstab

pd.crosstab(index= df['condition'],
            columns= df['grade'],
            values= df['price'],
            aggfunc= 'median')
```
