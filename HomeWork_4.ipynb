## 4.3 Работа с датасетом "house price"
### 3.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(15, 12))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

**Коэффицинты корреляции**
* до 0,2 - Очень слабая
* до 0,5 - Слабая
* до 0,7 - Средняя
* до 0,9 - Высокая
* свыше 0,9 - Очень высокая корреляция

**По тепловой карте наблюдается корреляция между стоимостью недвижимости и:**
* **grade** - оценкой дизайна и конструкции здания (**r = 0.7** - средняя корреляция);
* **sqft_living** - квадратурой жилой площади (**r = 0.7** - средняя корреляция);
* **sqft_above** - квадратура выше земли (**r = 0.6** - средняя)
* **bathrooms** - количество ванных комнат (**r = 0.5** - слабая)
* **view** - оценка, насколько хороший вид (**r = 0.4** - слабая)

#### **Диаграмма 1. Влияние оценки дизайна и конструкции здания (grade) на стоимость недвижимости (price)**
* **grade** - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна (r = 0.7 - средняя корреляция)

```python
plt.figure(figsize=(30, 18))

sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(x=df['grade'], y=df['price_1000'])

plt.title('Стоимость домов от оценки')
plt.xlabel('Оценка')
plt.ylabel('Стоимость')

plt.yticks(range(0, 5500, 500));
```

### Вывод :
* По шкале от 1 до 13 "оценка дизайна и конструкции здания" (1-3 плохо, 7 - средне, 11-13 - высоко):
* наибольшая стоимость у домов с высокой оценкой ("13"),
* наименьшая стоимость у домов с низкой оценкой ("1"),
* с увеличением оценки увеличивается стоимость.
* Соответственно, чем выше оценка, тем выше стоимость дома.
